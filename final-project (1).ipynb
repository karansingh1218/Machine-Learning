## Check All of the Training Data for missing values and general behavior of the sensors, since all of the data is fragmented

```python
logsTrainList
```

```python
def missing_data_parser(filename, filepath):
    
    log = filepath + filename

    df = pd.read_csv(log)
    
    var = df.isnull().sum().tolist()
    
    return var

def missing_data_aggregator(trainList, filepath):
    
    summary = []
    
    for i in range(len(trainList)):
        
        result = missing_data_parser(trainList[i], filepath)
        
        summary.append(result)    
    
    return summary
    
```

```python
result = missing_data_aggregator(logsTrainList,trainLogPath)
```

```python
column_names = ['sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5','sensor_6', 'sensor_7',
            'sensor_8', 'sensor_9', 'sensor_10']
result_df = pd.DataFrame(result, columns= column_names)
```

```python
def create_features(index, train_sel=1):
     
    ser = pd.Series()
    dir = "/kaggle/input/predict-volcanic-eruptions-ingv-oe/"
    if(train_sel ==1):
        subdir = "train/"
        segmentid = train.loc[index]["segment_id"]
        ttr = train.loc[index]["time_to_eruption"]
        
    else:
        subdir = "test/"
        segmentid = test.loc[index]["segment_id"]
        ttr = test.loc[index]["time_to_eruption"]
        
        
    log = dir + subdir + str(segmentid) + ".csv"
    df = pd.read_csv(log)
    
    df.interpolate(inplace=True)
    df.fillna(0,inplace=True)
    
    ser["segment_id"] = segmentid
    ser["time_to_eruption"] = ttr
    for sensor in df.columns:
        
        
        ser[sensor+"_max"] = df[sensor].max()
        ser[sensor+"_min"] = df[sensor].min()
        ser[sensor+"_sum"] = df[sensor].sum()
        ser[sensor+ "_mean"] = df[sensor].mean()
        ser[sensor+ "_mode"] = df[sensor].mode()
        ser[sensor+ "_std"] = df[sensor].std()
        ser[sensor+ "_var"] =  df[sensor].var()
        ser[sensor+ "_sem"] =  df[sensor].sem()
        ser[sensor+ "_skew"] =  df[sensor].skew()
        ser[sensor+  "_kurt"] = df[sensor].kurt()
        ser[sensor+  "_HQ"] = df[sensor].quantile(0.95)
        ser[sensor+ "_LQ"] =  df[sensor].quantile(0.05)
        ser[sensor+ "_Q1"] =   df.quantile(0.25)
        ser[sensor+  "_Q3"] =  df.quantile(0.75)
        ser[sensor+  "_IQR"] = ser[sensor+  "_Q3"] - ser[sensor+ "_Q1"]
    
    return ser
  
```

```python
params = {
    # Parameters that we are going to tune.
    'max_depth':6,
    'min_child_weight': 1,
    'eta':.3,
    'subsample': 1,
    'colsample_bytree': 1,
    # Other parameters
    'objective':'reg:linear',
}
params['eval_metric'] = "mae"
num_boost_round = 999

model = xg.train(
    params,
    dtrain,
    num_boost_round=num_boost_round,
    evals=[(dtest, "Test")],
    early_stopping_rounds=10
)
print("Best MAE: {:.2f} with {} rounds".format(
                 model.best_score,
                 model.best_iteration+1))

```

```python
cv_results = xg.cv(
    params,
    dtrain,
    num_boost_round=num_boost_round,
    seed=42,
    nfold=5,
    metrics={'mae'},
    early_stopping_rounds=10
)
cv_results
cv_results['test-mae-mean'].min()
```

```python
gridsearch_params = [
    (max_depth, min_child_weight)
    for max_depth in range(9,12)
    for min_child_weight in range(5,8)
]

# Define initial best params and MAE
min_mae = float("Inf")
best_params = None
for max_depth, min_child_weight in gridsearch_params:
    print("CV with max_depth={}, min_child_weight={}".format(
                             max_depth,
                             min_child_weight))
    # Update our parameters
    params['max_depth'] = max_depth
    params['min_child_weight'] = min_child_weight
    # Run CV
    cv_results = xg.cv(
        params,
        dtrain,
        num_boost_round=num_boost_round,
        seed=42,
        nfold=5,
        metrics={'mae'},
        early_stopping_rounds=10
    )
    # Update best MAE
    mean_mae = cv_results['test-mae-mean'].min()
    boost_rounds = cv_results['test-mae-mean'].argmin()
    print("\tMAE {} for {} rounds".format(mean_mae, boost_rounds))
    if mean_mae < min_mae:
        min_mae = mean_mae
        best_params = (max_depth,min_child_weight)
print("Best params: {}, {}, MAE: {}".format(best_params[0], best_params[1], min_mae))
```

```python
params['max_depth'] = 10
params['min_child_weight'] = 7

gridsearch_params = [
    (subsample, colsample)
    for subsample in [i/10. for i in range(7,11)]
    for colsample in [i/10. for i in range(7,11)]
]

min_mae = float("Inf")
best_params = None
# We start by the largest values and go down to the smallest
for subsample, colsample in reversed(gridsearch_params):
    print("CV with subsample={}, colsample={}".format(
                             subsample,
                             colsample))
    # We update our parameters
    params['subsample'] = subsample
    params['colsample_bytree'] = colsample
    # Run CV
    cv_results = xg.cv(
        params,
        dtrain,
        num_boost_round=num_boost_round,
        seed=42,
        nfold=5,
        metrics={'mae'},
        early_stopping_rounds=10
    )
    # Update best score
    mean_mae = cv_results['test-mae-mean'].min()
    boost_rounds = cv_results['test-mae-mean'].argmin()
    print("\tMAE {} for {} rounds".format(mean_mae, boost_rounds))
    if mean_mae < min_mae:
        min_mae = mean_mae
        best_params = (subsample,colsample)
print("Best params: {}, {}, MAE: {}".format(best_params[0], best_params[1], min_mae))
```
